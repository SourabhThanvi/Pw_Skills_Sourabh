## 12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

```python
import pandas as pd
```

```python
data = {'Name':['Sanju', 'KK', 'Manju', 'SAA', 'SRK'],
        'Marks': [22, 34, 45, 67, 56],
        'CGPA':[4, 5, 5.5, 7, 6.2]}
```

```python
df = pd.DataFrame(data)
```

```python
df.cov(numeric_only = True) # It does not standardize the result.
```

```python
df.corr(numeric_only = True) # It does standardize the result. and it ranges from -1 to +1.
```
