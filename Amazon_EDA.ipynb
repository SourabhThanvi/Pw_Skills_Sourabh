## Correcting the datatype

```python
df_a['discounted_price'].sample(5)
```

```python
df_a['actual_price'].sample(5)
```

```python
lst = ['actual_price', 'discounted_price']

def convert_to_num():
  for i in lst:
    df_a[i] = df_a[i].str.replace('₹','').str.replace(',','')
    df_a[i] = df_a[i].astype(float)
```

```python
convert_to_num()
```

```python
df_a.drop(index=1279, inplace=True)
```

```python
df_a['rating'] = df_a['rating'].astype(float)
```

```python
df_a['rating_count'].unique()
```

```python
df_a['rating_count'] = df_a['rating_count'].str.replace(',', '').astype(int)
```

## 3. What is the distribution of discounted prices vs. actual prices?

```python
df_a.columns
```

```python
sns.kdeplot(data=df_a, x=df_a['discounted_price'], fill=True, color='blue', alpha=0.5, label='Discounted Price')
sns.kdeplot(data=df_a, x=df_a['actual_price'], fill=True, color='red', alpha=0.5, label='Actual Price')
plt.xlabel('Discounted_price VS Actual_price')
plt.legend()
plt.show()

# discounted price has less varaince where as in comparsion to discounted price , actual price has more varaince
```
