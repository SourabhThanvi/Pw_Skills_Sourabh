```python
# Deep copy and shallow copy.

# Shallow Copy
x = np.array([1,2,3,4])
y = x
```

```python
# Changes in y will also reflect in x, because they both are pointing towards same memory address.
# This is called shallow copy.

print(y)
print(x)
```

```python
y[1] = 4
```

```python
print('Value at y', y)
print('Value at x', x)
print(id(y))
print(id(x))
```

```python
# Deep Copy

x = np.array([1,2,3,4])

y = x.copy()
```

```python
# Changes in y will not refelect in x varaible, because now we have created a copy of x and y is pointing towards a new memory address.

print(y)
print(x)
```

```python
y[1] = 4
```

```python
print('Value at y', y)
print('Value at x', x)
print(id(y))
print(id(x))
```

### 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

```python
date = pd.date_range('2024-09-01', periods=50)

temperature = np.random.uniform(20, 41, 50)
temperature = np.round(temperature, 2)

humidity = np.random.uniform(18, 93, 50)
humidity = np.round(humidity, 2)
```

```python
df14 = pd.DataFrame()

df14['Date'] = date
df14['Temperature'] = temperature
df14['Humidity'] = humidity
```

### 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
x = np.random.randn(100)
y = np.random.randn(100)
```

```python
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

quadrants = [get_quadrant(x, y) for x, y in zip(x, y)]
```

```python
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})
```

```python
sns.relplot(x=data['x'], y=data['y'], kind='scatter', hue = data['Quadrant'] )

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Quadrant-Wise Scatter Plot')
```
